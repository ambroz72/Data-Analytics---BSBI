```python
qt_no=[]
qt_o=[]
qt_d=[]
qt_c=[]
for colum in data:
  if data[colum].dtype == 'object':
    qt_o.append(colum)
  elif data[colum].dtype == 'int64':
    qt_no.append(colum)
  elif data[colum].dtype == 'float64':
    qt_d.append(colum)
  else:
    qt_c.append(colum)
    print("Qualitative - Nominal:", qt_no)
print("Qualitative - Ordinal:", qt_o)
print("Quantitative - Discrete:", qt_d)
print("Quantitative - Continuous:", qt_c)
```

```python
# missing values
print("Missing values in each column:")
print(data.isnull().sum())
```

```python
# Handling missing values
# Option 1: Fill missing values with mean/median for numerical columns
for column in data.select_dtypes(include=[np.number]).columns:
    if data[column].isnull().sum() > 0:
        data[column].fillna(data[column].mean(), inplace=True)  # Replace with `median()` if preferred
```

```python
# Option 2: Drop rows with missing values
data.dropna(inplace=True)
```

```python
data
```

```python
# Option 3: Drop columns with too many missing values
threshold = 0.5 * len(data)  # 50% of rows as threshold
data = data.loc[:, data.isnull().sum() < threshold]
```

```python
# Convert categorical variables to category type
for column in data.select_dtypes(include=['object']).columns:
    data[column] = data[column].astype('category')
```

```python
data
```

```python
# Display updated data types
print("\nUpdated data types:")
print(data.dtypes)
```

```python
# 3. Create new columns if necessary
# Example: Extract systolic and diastolic blood pressure from a single column
# Assuming column name is 'BloodPressure' in the format '120/80'
if 'BloodPressure' in data.columns:
    data[['Systolic', 'Diastolic']] = data['BloodPressure'].str.split('/', expand=True).astype(float)
    data.drop(columns=['BloodPressure'], inplace=True)  # Optional: drop the original column

# Preview cleaned data
print("\nPreview of cleaned data:")
print(data.head())
```

```python
data.info()
```

```python
# Check for duplicates
duplicate_count = data.duplicated().sum()
print(f"Number of duplicate rows: {duplicate_count}")

# Optional: Handle duplicates
if duplicate_count > 0:
    # Drop duplicate rows
    data = data.drop_duplicates()
    print("Duplicate rows removed.")
```

```python
# List of numerical variables to analyze
numerical_vars = ['Age', 'Sleep Duration', 'Physical Activity Level', 'Daily Steps', 'Heart Rate']

# Split Blood Pressure into Systolic and Diastolic for analysis
data[['Systolic BP', 'Diastolic BP']] = data['blood.pressure'].str.split('/', expand=True).astype(int)

# Add Blood Pressure components to the numerical variables list
numerical_vars.extend(['Systolic BP', 'Diastolic BP'])

# Visualize distributions and box plots for numerical variables
for var in numerical_vars:
    if var in data.columns:
        plt.figure(figsize=(12, 6))

        # Histogram
        plt.subplot(1, 2, 1)
        sns.histplot(data[var], kde=True, bins=30, color='blue')
        plt.title(f'Distribution of {var}')
        plt.xlabel(var)
        plt.ylabel('Frequency')

        # Box Plot
        plt.subplot(1, 2, 2)
        sns.boxplot(x=data[var], color='orange')
        plt.title(f'Box Plot of {var}')
        plt.xlabel(var)

        plt.tight_layout()
        plt.show()
```

```python
# --- Univariate Analysis for Categorical Variables ---
# Define the list of key categorical variables
categorical_vars = ['Gender', 'Occupation', 'BMI Category', 'Quality of Sleep', 'Stress Level', 'Sleep Disorder']
```

```python
# Visualize distributions using bar charts
for var in categorical_vars:
    if var in data.columns:
        plt.figure(figsize=(8, 4))
        sns.countplot(x=data[var], palette='Set2', order=data[var].value_counts().index)
        plt.title(f'Distribution of {var}')
        plt.xlabel(var)
        plt.ylabel('Count')
        plt.xticks(rotation=45)
        plt.show()
```

```python
# --- Identify Outliers in Numerical Data ---
# Using IQR method to detect outliers
for var in numerical_vars:
    if var in data.columns:
        Q1 = data[var].quantile(0.25)
        Q3 = data[var].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        outliers = data[(data[var] < lower_bound) | (data[var] > upper_bound)]
        print(f"{var}: Found {len(outliers)} outliers")
```

```python
# Check if 'Gender' column exists
if 'gender' in data.columns:
    # Calculate percentage distribution
    gender_counts = data['gender'].value_counts(normalize=True) * 100

    print("Gender Distribution (Percentage):")
    print(gender_counts)

    # Visualize the distribution
    plt.figure(figsize=(6, 4))
    sns.barplot(x=gender_counts.index, y=gender_counts.values, palette="Set2")
    plt.title("Gender Distribution")
    plt.xlabel("gender")
    plt.ylabel("Percentage")
    plt.show()
else:
    print("The column 'Gender' does not exist in the dataset.")
```

```python
# Calculate IQR for 'Sleep Duration' and 'Daily Steps'
Q1_sleep = data['sleep.duration'].quantile(0.25)
Q3_sleep = data['sleep.duration'].quantile(0.75)
IQR_sleep = Q3_sleep - Q1_sleep

Q1_steps = data['daily.steps'].quantile(0.25)
Q3_steps = data['daily.steps'].quantile(0.75)
IQR_steps = Q3_steps - Q1_steps

# Determine outlier boundaries
lower_bound_sleep = Q1_sleep - 1.5 * IQR_sleep
upper_bound_sleep = Q3_sleep + 1.5 * IQR_sleep

lower_bound_steps = Q1_steps - 1.5 * IQR_steps
upper_bound_steps = Q3_steps + 1.5 * IQR_steps
# Identify outliers
outliers_sleep = data[(data['sleep.duration'] < lower_bound_sleep) | (data['sleep.duration'] > upper_bound_sleep)]
outliers_steps = data[(data['daily.steps'] < lower_bound_steps) | (data['daily.steps'] > upper_bound_steps)]

# Visualize outliers using box plots
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
sns.boxplot(x=data['sleep.duration'])
plt.title('Box Plot of Sleep Duration')

plt.subplot(1, 2, 2)
sns.boxplot(x=data['daily.steps'])
plt.title('Box Plot of Daily Steps')

plt.tight_layout()
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler

# Assuming your data is in a DataFrame named 'df'

# Select columns for normalization
columns_to_normalize = ['age', 'sleep.duration', 'physical.activity.level']

# Min-Max Scaling
scaler = MinMaxScaler()
data_minmax = data.copy()
data_minmax[columns_to_normalize] = scaler.fit_transform(data[columns_to_normalize])

# Z-score Normalization
scaler = StandardScaler()
data_zscore = data.copy()
data_zscore[columns_to_normalize] = scaler.fit_transform(data[columns_to_normalize])

# Concatenate original and normalized dataframes
data_combined = pd.concat([data, data_minmax, data_zscore], axis=1)

# Print the first few rows of the combined dataframe
print(data_combined.head())
```
